```python
import pandas as pd
import numpy as np

# Load cleaned dataset from Day 6
df = pd.read_csv("../data/cleaned_churn.csv")

print("\n--- DATA BEFORE FEATURE ENGINEERING ---")
print(df.head())

# 1️⃣ tenure_months (explicit feature)
df["tenure_months"] = df["tenure"]

# Replace tenure 0 with 1 to avoid division by zero
df["tenure_months"] = df["tenure_months"].replace(0, 1)

# 2️⃣ average_monthly_usage
# (TotalCharges divided by tenure)
df["average_monthly_usage"] = df["TotalCharges"] / df["tenure_months"]

# 3️⃣ charges_per_month
# (MonthlyCharges compared to tenure)
df["charges_per_month"] = df["MonthlyCharges"] / df["tenure_months"]

print("\n--- DATA AFTER FEATURE ENGINEERING ---")
print(df[["tenure", "tenure_months", "average_monthly_usage", "charges_per_month"]].head())

# Save feature-engineered dataset
df.to_csv("../data/feature_engineered_churn.csv", index=False)

print("\nDAY 8 COMPLETED: Feature engineering successful")
```
